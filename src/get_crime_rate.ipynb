```python
# Import dependencies here
# this application uses BeautifulSoup to scrape web data
# Scrapes crime rates data from crimerate.com

from bs4 import BeautifulSoup
import requests
import urllib3
import re
import pandas as pd
import time
```

```python
zipcodes_csv = pd.read_csv('../data/zipcodes.csv')
zipcodes_lst = zipcodes_csv['zip'].values
```

```python
url = 'http://crimerate.com/index.php?st='
http = urllib3.PoolManager()
f = open('../data/crime_raw_report.txt','w')

counter = 0
for z in zipcodes_lst:
    print(f"{url}{z}")
    query_url = f"{url}{z}"
    response = http.request('GET', query_url)
    soup = BeautifulSoup(response.data)
    for i in soup.get_text().split('\n'):
        if re.match('Crime rates in', i):
            str = f"Zip{z}:{i}"
            print(str)
            print(str, file=f)    
    counter += 1
    if counter % 10 == 0:
        print(f'{counter} records have been fetched so far')
        time.sleep(2)
        
f.close()
```
