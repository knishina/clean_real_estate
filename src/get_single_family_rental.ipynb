### Find rental cost of single family homes with given zip codes.

```python
import json
import requests
import pandas as pd
import numpy as np
import csv
import time
import quandl
import urllib
quandl.ApiConfig.api_key = ''
```

```python
datafile1 = "../data/Single_Family_Home_Cost.csv"
df1 = pd.read_csv(datafile1)
df1.head()
```

```python
rentals = []
for x in df1["Zip Code"]:
    input = f"ZILLOW/Z{x}_MRPAH"
    try:
        mydata = (quandl.get(input))["Value"][-1]
        rentals.append(mydata)
    except:
        rentals.append(0)
print (rentals)
```

```python
df1["Rental Prices"] = rentals
df1.head(32)
```

```python
# This is from zillow's csv.  There are a couple additional data points that were not part of the quandl api.

datafile2 = "../data/Rental1.csv"
df2 = pd.read_csv(datafile2)
df2 = df2[["RegionName", "2018-02"]]
df2 = df2.rename(columns= {"RegionName": "Zip Code"})
df2.head()
```

```python
df3 = pd.merge(df1, df2, on = "Zip Code", how ="left")
df3["2018-02"] = df3["2018-02"].fillna(0)
df3 = df3.rename(columns= {"Rental Prices": "Quandl"})
df3.head()
```

```python
new_list = []
for x in range(len(df3)):
    if df3["Quandl"][x] == 0 and df3["2018-02"][x] == 0:
        new_list.append(0)
    elif df3["Quandl"][x] == 0 and df3["2018-02"][x] != 0:
        new_list.append(df3["2018-02"][x])
    elif df3["2018-02"][x] == 0 and df3["Quandl"][x] != 0:
        new_list.append(df3["Quandl"][x])
    else:
        average = round((df3["2018-02"][x] + df3["Quandl"][x])/ 2, 2)
        new_list.append(average)
print (new_list)
len(new_list)
```

```python
df3["Rental Prices"] = new_list
df3 = df3[["Cost of Single Family Home", "Zip Code", "Rental Prices"]]
df3.head()
```

```python
stuff = []
for x in df3["Zip Code"]:
    url = f"http://www.city-data.com/zips/{x}.html"
    stuff.append(url)

rental_value = []
for y in stuff:
    time.sleep(0.5)
    try:
        response = urllib.request.urlopen(y).read()
        response = response.decode("utf-8")
        t = "Median gross rent in 2016:</b> $"
        aa = response.split(t)
        bb = aa[1].split(".")
        rent = bb[0]
        rental_value.append(rent)
    except:
        rental_value.append(0)
```

```python
print (rental_value)
print (len(rental_value))
```

```python
df3["Rental Value 2016"] = rental_value
df3["Rental Value 2016"] = df3["Rental Value 2016"].replace(",", "")
df3["Rental Value 2016"] = df3["Rental Value 2016"].astype("int")
df3.to_csv("../data/Single_Family_Rental.csv")
df3
```
